## üì¶ Imports

```python
import os
import cv2
import torch
import numpy as np
import pandas as pd
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader, Subset
import torch.nn as nn
from torchvision.models import resnet18
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from tqdm import tqdm
from collections import Counter
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedShuffleSplit
from torchvision.models import resnet34
```

## ‚öôÔ∏è Configura√ß√µes

```python
# =============================
# Configura√ß√µes
# =============================
BATCH_SIZE = 1
EPOCHS = 50
LR = 1e-4
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
FRAME_DIR = 'preprocessed_frames'
NUM_FRAMES = 16

# =============================
# Transforma√ß√µes
# =============================
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])
```

## üß† Treino do Modelo

```python
# Distribui√ß√£o de classes: ({0: 22, 1: 13, 2: 21, 3: 4})

# =============================
# Pr√©-processamento
# =============================

# torch.Tensor com shape [16, 3, 224, 224]
# 16 frames, 3 channels (RGB), cada frame com resolu√ß√£o 224√ó224 pixels

def preprocess_and_save_frames(csv_path, video_dir):
    df = pd.read_csv(csv_path)
    os.makedirs(FRAME_DIR, exist_ok=True)

    for _, row in tqdm(df.iterrows(), total=len(df)):
        video_id = row['VIDEO']
        video_path = os.path.join(video_dir, video_id + '.mp4')
        save_path = os.path.join(FRAME_DIR, video_id + '.pt')

        if os.path.exists(save_path):
            continue

        cap = cv2.VideoCapture(video_path)
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        if total_frames == 0:
            print(f"[ERRO] Sem frames: {video_id}")
            continue

        step = max(total_frames // NUM_FRAMES, 1)
        frames = []

        for i in range(NUM_FRAMES):
            cap.set(cv2.CAP_PROP_POS_FRAMES, i * step)
            ret, frame = cap.read()
            if not ret or frame is None:
                continue
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            tensor = transform(frame_rgb).float()
            frames.append(tensor)

        cap.release()

        if len(frames) == 0:
            print(f"[ERRO] Nenhum frame lido: {video_id}")
            continue

        if len(frames) < NUM_FRAMES:
            pad_frames = [torch.zeros_like(frames[0]) for _ in range(NUM_FRAMES - len(frames))]
            frames.extend(pad_frames)

        video_tensor = torch.stack(frames[:NUM_FRAMES])
        torch.save(video_tensor, save_path)

# =============================
# Dataset
# =============================
class PreprocessedVideoDataset(Dataset):
    def __init__(self, csv_file):
        self.data = pd.read_csv(csv_file)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        video_id = row['VIDEO']
        tensor_path = os.path.join(FRAME_DIR, video_id + '.pt')
        frames = torch.load(tensor_path).float()

        grs = int(row['GLOBA_RATING_SCORE'])
        if 8 <= grs <= 15:
            label = 0
        elif 16 <= grs <= 23:
            label = 1
        elif 24 <= grs <= 31:
            label = 2
        elif 32 <= grs <= 40:
            label = 3
        else:
            raise ValueError(f"GRS fora do intervalo esperado (8‚Äì40): {grs}")

        return frames, label

# =============================
# Modelo CNN + LSTM (Classifica√ß√£o)
# =============================
class CNNEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        base = resnet34(pretrained=True)
        self.feature_extractor = nn.Sequential(*list(base.children())[:-1])

    def forward(self, x):
        x = self.feature_extractor(x)
        return x.view(x.size(0), -1)

class CNN_LSTM_Model(nn.Module):
    def __init__(self, hidden_dim=128, num_classes=4):
        super().__init__()
        self.cnn = CNNEncoder()
        self.lstm = nn.LSTM(input_size=512, hidden_size=hidden_dim, batch_first=True, dropout=0.3)
        self.fc = nn.Linear(hidden_dim, num_classes)

    def forward(self, x):
        B, T, C, H, W = x.size()
        x = x.view(B * T, C, H, W)
        features = self.cnn(x)
        features = features.view(B, T, -1)
        _, (hn, _) = self.lstm(features)
        return self.fc(hn.squeeze(0))

# =============================
# Treino e avalia√ß√£o
# =============================
def train_model(model, train_dl, val_dl):
    model.to(DEVICE)
    optimizer = torch.optim.Adam(model.parameters(), lr=LR)

    # Calcular pesos por classe
    all_labels = []
    for _, label in train_dl.dataset:
        all_labels.append(label)
    counts = Counter(all_labels)
    total = sum(counts.values())
    weights = [total / counts.get(i, 1) for i in range(4)]
    weights = torch.tensor(weights, dtype=torch.float32).to(DEVICE)
    criterion = nn.CrossEntropyLoss(weight=weights)

    loss_history = []

    for epoch in range(EPOCHS):
        model.train()
        total_loss, correct, total = 0, 0, 0
        for videos, labels in train_dl:
            videos, labels = videos.to(DEVICE), labels.to(DEVICE)
            outputs = model(videos)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()

            total_loss += loss.item()
            correct += (outputs.argmax(1) == labels).sum().item()
            total += labels.size(0)

        acc = correct / total
        avg_loss = total_loss / len(train_dl)
        loss_history.append(avg_loss)
        print(f"Epoch {epoch+1}/{EPOCHS} | Loss: {avg_loss:.4f} | Acc: {acc:.4f}")

    # Plotar a curva de perda
    plt.figure(figsize=(8, 4))
    plt.plot(range(1, EPOCHS + 1), loss_history, marker='o')
    plt.title('Loss por √âpoca')
    plt.xlabel('√âpoca')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.show()

    return model

def evaluate(model, test_dl):
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for videos, labels in test_dl:
            videos = videos.to(DEVICE)
            outputs = model(videos)
            preds = outputs.argmax(1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(labels.numpy())

    print("\nüîç Resultados da Classifica√ß√£o:")
    for true, pred in zip(all_labels, all_preds):
        print(f"Real: {true} | Previsto: {pred}")

    print("\nRelat√≥rio de Classifica√ß√£o:")
    print(classification_report(all_labels, all_preds))
    print("Matriz de Confus√£o:")
    print(confusion_matrix(all_labels, all_preds))

# =============================
# Execu√ß√£o
# =============================

csv_path = 'filtered_osats_avg.csv'
video_dir = 'Package01'

#print("üîß Pr√©-processamento dos v√≠deos...")
#preprocess_and_save_frames(csv_path, video_dir)

print("üì¶ A carregar dataset j√° processado...")
full_dataset = PreprocessedVideoDataset(csv_path)
dataset = Subset(full_dataset, range(60))
#dataset = full_dataset
# train_data, test_data = train_test_split(dataset, test_size=0.3, random_state=42)
all_labels = [full_dataset[i][1] for i in range(len(full_dataset))]
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=42)

for train_idx, test_idx in sss.split(np.zeros(len(all_labels)), all_labels):
    train_data = Subset(full_dataset, train_idx)
    test_data = Subset(full_dataset, test_idx)

train_dl = DataLoader(train_data, batch_size=BATCH_SIZE, shuffle=True)
test_dl = DataLoader(test_data, batch_size=BATCH_SIZE)

print("üöÄ A treinar modelo...")
model = CNN_LSTM_Model()
model = train_model(model, train_dl, test_dl)
evaluate(model, test_dl)
```

## üçÄ Previs√µes

```python
# =============================
# Infer√™ncia em v√≠deo novo
# =============================
def inferir_video(model, video_path):
    model.eval()
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if total_frames == 0:
        print("[ERRO] O v√≠deo est√° vazio.")
        return

    step = max(total_frames // NUM_FRAMES, 1)
    frames = []

    for i in range(NUM_FRAMES):
        cap.set(cv2.CAP_PROP_POS_FRAMES, i * step)
        ret, frame = cap.read()
        if not ret or frame is None:
            continue
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        tensor = transform(frame_rgb).float()
        frames.append(tensor)

    cap.release()

    if len(frames) == 0:
        print("[ERRO] Nenhum frame v√°lido lido.")
        return

    if len(frames) < NUM_FRAMES:
        pad_frames = [torch.zeros_like(frames[0]) for _ in range(NUM_FRAMES - len(frames))]
        frames.extend(pad_frames)

    video_tensor = torch.stack(frames[:NUM_FRAMES]).unsqueeze(0).to(DEVICE)
    with torch.no_grad():
        output = model(video_tensor)
        predicted_class = output.argmax(1).item()

    faixa = ["8‚Äì15 (Novato)", "16‚Äì23 (Interm√©dio)", "24‚Äì31 (Proficiente)", "32‚Äì40 (Especialista)"]
    print(f"\nüé• {os.path.basename(video_path)} ‚Üí Classe prevista: {predicted_class} | Faixa GRS: {faixa[predicted_class]}")

inferir_video(model, "Package03/F75X.mp4")
```
